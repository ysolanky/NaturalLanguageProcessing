```python
import nltk 
```

```python
text = nltk.word_tokenize("They wind back the clock, while we chase after the wind")
```

```python
nltk.pos_tag(text)
```

```python
# They is a PRP = Personal pronoun
# Wind is both a present tense Verb and a Noun. Might be because of different pronunciations. 
# back is a RB - adverb
# the is a DT - determiner
# clock is a NN  - Noun
# While is IN - preposition
# we is PRP - Personal pronoun
# chase is VBP - present tense verb
# after is IN - preposition
```
